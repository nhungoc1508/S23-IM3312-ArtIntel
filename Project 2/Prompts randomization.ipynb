```python
import pandas as pd
import numpy as np
import random
import json
```

```python
prompts_df = pd.read_csv('prompts_no_style.csv')
print(prompts_df.shape)
prompts_df.head()
```

```python
def get_timestamp(time):
  minute = int(time[0])
  seconds = int(time[2:])
  return minute * 60 + seconds
```

```python
prompts_df['start'] = prompts_df.apply(lambda x: get_timestamp(x.timestamp_start), axis=1)
prompts_df['end'] = prompts_df.apply(lambda x: get_timestamp(x.timestamp_end), axis=1)
```

```python
prompts_df.tail()
```

```python
with open('styles.json') as json_file:
  data = json.load(json_file)
```

```python
data.keys()
```

```python
styles = [style for _, value in data.items() for style in value]
len(styles)
```

```python
def get_style(desc):
  return '{0}, {1}, {2}, {3}, {4}'.format(desc.replace('.', ''), *random.sample(styles, 4))
```

```python
prompts_df['prompt'] = prompts_df.apply(lambda x: get_style(x.description), axis=1)
```

```python
for i in range(4):
  print(prompts_df.iloc[i].prompt)
```
